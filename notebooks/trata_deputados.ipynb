```python
#Importa as bibliotecas pandas e abravia ela para pd e a biblioteca json
import pandas as pd
import json

#Carrega o arquivo como um dicionario python
with open('deputados_brutos.json', 'r', encoding='utf-8') as f:
    conteudo = json.load(f)

#Achata os dados que estao dentro da chave dados
df = pd.json_normalize(conteudo['dados'])

#Identifica as variaveis do email nulo e as substitui
df['email'] = df['email'].fillna('Não informado')

#Imprime as colunas nome e email para verificação
print(df[['nome', 'email']])

#Converte o Dataframe para uma lista de dicionario python
dados_finais = df.to_dict(orient='records')

#Salva usando a biblioteca json
with open('dados_atualizados.json', 'w', encoding='utf-8') as f:
    json.dump(dados_finais, f, indent=4, ensure_ascii=False)
```
