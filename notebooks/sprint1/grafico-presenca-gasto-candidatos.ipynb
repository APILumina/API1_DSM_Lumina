```
#importação
import pandas as pd
import matplotlib.pyplot as plt
#leitura
df1 = pd.read_csv(
    'https://raw.githubusercontent.com/APILumina/API1_DSM_Lumina/feature/codigos-graficos/data/deputados/ranking_gastos_deputados.csv',
    sep=';',
    low_memory=False #evitar erros colunas grandes
)

df1.groupby('nome')['valor'].sum().sort_values().tail(20).plot(kind='barh', figsize=(10,10))

plt.title('Ranking de Gastos Deputados')
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt


#leitura
df1 = pd.read_csv(
    'https://raw.githubusercontent.com/APILumina/API1_DSM_Lumina/refs/heads/feature/codigos-graficos/data/deputados/ranking_presencas_deputados.csv',

    sep=';',
    low_memory=False #evitar erros colunas grandes
)


df1.groupby('nome')['total_presencas'].sum().sort_values().tail(20).plot(kind='barh', figsize=(10,10))

plt.title('Ranking de Presenças por Deputados')
plt.show()
```
