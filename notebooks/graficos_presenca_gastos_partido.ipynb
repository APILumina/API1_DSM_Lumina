```python
#Importação das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive as drive

#Local para pegar os arquivos
url_df1 = 'https://raw.githubusercontent.com/APILumina/API1_DSM_Lumina/feature/codigos-graficos/data/deputados/ranking_presencas_deputados.csv'
url_df2 = 'https://raw.githubusercontent.com/APILumina/API1_DSM_Lumina/feature/codigos-graficos/data/deputados/deputados_detalhes_foto.csv'

# Otimização da visualização de tabelas e gráficos
low_memory=False
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90

#Criação dos dataFrame

df1 = pd.read_csv(url_df1, sep=';')
df2 = pd.read_csv(url_df2, sep=';')

df_final = pd.merge(df1, df2, left_on='nome', right_on='nome_eleitoral')

#Criação das tabelas
df_final.groupby('partido')['total_presencas'].sum().sort_values().tail(20).plot(kind='barh', figsize=(10,10))
plt.title('Ranking de Presença por Partido')
```
