```python
import requests
import csv

#1. Atraves do endpoint da api, pegamos os dados dos deputados em ordem e somente na legislatura atual(57)
url = "https://dadosabertos.camara.leg.br/api/v2/deputados?ordem=ASC&ordenarPor=nome"
resposta = requests.get(url, params={'idLegislatura': 57})

#2.Podemos prosseguir apenas se code == 200, pois significa que deu tudo certo
if resposta.status_code == 200:
    dados_brutos = resposta.json()
    #ele retorna um dicionário, queremos somente a chave "dados"
    lista_original = dados_brutos['dados']

    lista_limpa = []
    #3. Pegamos apenas id e nome, através das chaves e adicionamos à lista_limpa
    for dep in lista_original:
        dicionario_reduzido = {
            "id": dep['id'],
            "nome": dep['nome'],
        }
        lista_limpa.append(dicionario_reduzido)

    #4. Para criar arquivo csv
    with open("dep_id_nome.csv", "w", encoding="utf-8") as arquivo:
        #Define nome das colunas
        escritor = csv.DictWriter(arquivo, fieldnames=["id", "nome"], delimiter=';')
        escritor.writeheader()   #Escreve o cabeçalho
        escritor.writerows(lista_limpa) #Escreve os dados da lista

    print(f"Sucesso! Arquivo gerado com {len(lista_limpa)} deputados.")

else:
    print(f"Erro na conexão: {resposta.status_code}")
```
