Códigos para a criação dos Gráficos de Gastos e Presenças/Partido

```python
#Importação das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive as drive

#Local para pegar os arquivos
drive.mount('/content/drive')

# Otimização da visualização de tabelas e gráficos
low_memory=False
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90

#Criação dos dataFrame

df1 = pd.read_csv('/content/drive/MyDrive/Gráficos Gastos e Presença partido/ranking_presencas_deputados.csv', sep=';')
df2 = pd.read_csv('/content/drive/MyDrive/Gráficos Gastos e Presença partido/deputados_detalhes_foto.csv', sep=';')

df_final = pd.merge(df1, df2, on= 'nome_eleitoral')

#Criação das tabelas
df_final.groupby('partido')['total_presencas'].sum().sort_values().tail(20).plot(kind='barh', figsize=(10,10))
plt.title('Ranking de Presença por Partido')
```

```python
#Importação das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive as drive

#Local para pegar os arquivos
drive.mount('/content/drive')

# Otimização da visualização de tabelas e gráficos
low_memory=False
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90

#Criação dos dataFrame
df1 = pd.read_csv('/content/drive/MyDrive/Gráficos Gastos e Presença partido/ranking_gastos_deputados.csv', sep=';')
df2 = pd.read_csv('/content/drive/MyDrive/Gráficos Gastos e Presença partido/deputados_detalhes_foto.csv', sep=';')

df_final = pd.merge(df1, df2, on= 'nome_eleitoral')

#Criação das tabelas
df_final.groupby('partido')['valor'].sum().sort_values().tail(20).plot(kind='barh', figsize=(10,10))
plt.title('Ranking de Gastos por Partido')
```
