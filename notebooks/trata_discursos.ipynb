```python
import pandas as pd

# Carregar o arquivo
df = pd.read_csv('discursos_dep.csv', sep=';', encoding='utf-8-sig')

#Substitui as valores nulos/vazios para 'não disponiveis'
df['keywords'] = df['keywords'].fillna('Palavras-chave não disponiveis')
df['sumario'] = df['sumario'].fillna('Sumario não disponivel')

#df_limpo é o nome do seu DataFrame após as filtragens
df.to_csv('dados_deputados_limpos.csv',
                sep=';',
                index=False,
                encoding='utf-8-sig')
```
