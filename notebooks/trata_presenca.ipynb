```python
import pandas as pd
import json

with open('presencas_brutas_brasil_2025.json', 'r', encoding='utf-8') as f:
  conteudo_presencas = json.load(f)

#Achata os dados do arquivo de presenças em um novo DataFrame
df_presencas = pd.json_normalize(conteudo_presencas['dados'])

#Aplique as operações no novo DataFrame
df_presencas['nomeDeputado'] = df_presencas['nomeDeputado'].str.strip().str.upper()

#Faz as contagens de presença dos deputados
contagem = df_presencas['nomeDeputado'].value_counts()

print("Quantidade de sessões por deputado:")
print(contagem)
```
