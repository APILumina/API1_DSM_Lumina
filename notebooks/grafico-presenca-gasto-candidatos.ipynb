```
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive as drive

low_memory=False
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90

drive.mount('/content/drive')

df1 = pd.read_csv('/content/drive/MyDrive/ranking_presencas_deputados.csv', sep=';')
df1.groupby('nome')['total_presencas'].sum().sort_values().tail(20).plot(kind='barh', figsize=(10,10))
plt.title('Ranking de Presenças Deputados')
```
