# Tratamento Presenças


```python
import pandas as pd

# 1. Carregar o arquivo
df = pd.read_csv('radar_presencas.csv', sep=';', encoding='utf-8-sig')

# 2. Tratando valores nulos (NaN)
# Preenchemos campos vazios para não dar erro na visualização do seu grupo
df['data'] = df['data'].fillna('Data não informada')
df['tipo_evento'] = df['tipo_evento'].fillna('Evento não especificado')
df['situacao'] = df['situacao'].fillna('Status não informado')

# --- AGREGAÇÃO: Ranking de Presenças ---
# Usamos o .size() para contar quantas linhas (eventos) cada deputado tem.
# O reset_index(name='total_presencas') serve para dar um nome bonito à nova coluna de contagem.
ranking_presencas = df.groupby(['nome', 'uf']).size().sort_values(ascending=False).reset_index(name='total_presencas')

# 3. Salvar o arquivo principal limpo
df.to_csv('radar_presencas_limpo.csv',
          sep=';',
          index=False,
          encoding='utf-8-sig')

# 4. Salvar o Ranking de Assiduidade (quem mais aparece na Câmara)
ranking_presencas.to_csv('ranking_presencas_deputados.csv',
                         sep=';',
                         index=False,
                         encoding='utf-8-sig')

print("Tratamento de presenças concluído!")
print(f"O deputado mais ativo em eventos foi: {ranking_presencas.iloc[0]['nome']} com {ranking_presencas.iloc[0]['total_presencas']} presenças registradas.")
```

# Tratamento despesas


```python
import pandas as pd

# 1. Carregar o arquivo (usando suas configurações de separador e encoding)
df = pd.read_csv('radar_despesas.csv', sep=';', encoding='utf-8-sig')

# 2. Tratando valores nulos (NaN)
# Se a data estiver vazia, colocamos um aviso. Se o tipo sumiu, avisamos também.
df['data'] = df['data'].fillna('Data não informada')
df['tipo'] = df['tipo'].fillna('Tipo de despesa não especificado')

# 3. Garantindo que a coluna 'valor' seja numérica
# Às vezes a API manda o valor como string. Isso garante que vira número para somar.
df['valor'] = pd.to_numeric(df['valor'], errors='coerce').fillna(0)

# --- DICA EXTRA: AGREGAÇÃO (Isso vai brilhar no seu projeto) ---
# Vamos criar um resumo de quanto cada DEPUTADO gastou no total
resumo_gastos = df.groupby(['nome', 'uf'])['valor'].sum().sort_values(ascending=False).reset_index()

# 4. Salvar o arquivo limpo
df.to_csv('radar_despesas_limpo.csv',
          sep=';',
          index=False,
          encoding='utf-8-sig')

# 5. Salvar um arquivo só com o Ranking de Gastos (para facilitar a vida do seu grupo)
resumo_gastos.to_csv('ranking_gastos_deputados.csv',
                     sep=';',
                     index=False,
                     encoding='utf-8-sig')

print("Tratamento concluído!")
print(f"O deputado que mais gastou foi: {resumo_gastos.iloc[0]['nome']} com R$ {resumo_gastos.iloc[0]['valor']:.2f}")
```
